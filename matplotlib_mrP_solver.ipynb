```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.rand(len(x))
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(8,3))
# plt.plot(x,y,'o--',color='purple',lw=0.8,ms=5)
plt.plot(x,y,'o--',color='purple',lw=0.8,ms=5,label='Component 1')
plt.xlabel('time [s]')
plt.ylabel('voltage [V]')
# plt.legend()
# plt.legend(loc='upper right')
plt.legend(loc='upper right', fontsize=20)
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o',color='purple',lw=0.7,ms=5,label='Component 1')
plt.plot(x2,y2,'-',color='green',lw=0.7,ms=5,label='Component 2')
plt.xlabel('time [s]')
plt.ylabel('voltage [V]')
plt.legend(loc='upper right', fontsize=20)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o',color='purple',lw=0.7,ms=5,label='Data')
plt.plot(x2,y2,'-',color='green',lw=0.7,ms=5,label='Theory')
plt.xlabel('time [s]')
plt.ylabel('voltage [V]')
plt.ylim(top=1.5)
plt.legend(loc='upper right', fontsize=20, ncol=2)
```
