```python
# integral and derivatives
```

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y,x)
y_int = np.cumsum(y) * (x[1]-x[0])
```

```python
plt.plot(x,y)
plt.plot(x,dydx)
plt.plot(x,y_int)
```
